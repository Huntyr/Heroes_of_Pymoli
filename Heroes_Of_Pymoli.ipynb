```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "heroesofpymoli.csv"

# Read Purchasing File and store into Pandas data frame
hero_df = pd.read_csv(file_to_load)
```

```python
#PLAYER COUNT
#Get unique players
uni_player = hero_df["SN"].nunique()
uni_player_df = pd.DataFrame([uni_player], columns= ["Total Players"])
uni_player_df
```

```python
#Purchasing Analysis (Total)

  #Run basic calculations to obtain number of unique items, average price, # of Purchases, Total Revenue

uni_items = hero_df["Item ID"].nunique()
  #uni_items
ave_price = hero_df["Price"].mean()
  #ave_price
num_pur = len(hero_df)
  #num_pur
tot_rev = hero_df["Price"].sum()
  #tot_rev

  #Create a summary data frame to hold the results
PA_df = pd.DataFrame([[uni_items, ave_price, num_pur, tot_rev]], columns=["Unique Items", "Average Item Price", "Number of Purchases", "Total Revenue"])


  #Optional: give the displayed data cleaner formatting
  #Display the summary data frame
PA_df
```

```python
#Gender Demographics
len_gender = len(hero_df)
  
    #Percentage and Count of Male Players
m_count = (hero_df["Gender"] == "Male").sum()
m_per = m_count/len_gender * 100
#m_count
#m_per
  
    #Percentage and Count of Female Players
f_count = (hero_df["Gender"] == "Female").sum()
f_per = f_count/len_gender * 100
#f_count
#f_per  
    #Percentage and Count of Other / Non-Disclosed
o_count = (hero_df["Gender"] == "Other / Non-Disclosed").sum()
o_per = o_count/len_gender * 100
#o_count

    #Display table
gdem_df = pd.DataFrame({"Total Count": [m_count, f_count, o_count], "Total Percentage": [m_per, f_per, o_per]})
gdem_df
```

```python
#Purchasing Analysis (Gender)
    #Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
    
    #perchase count is m_count, f_count, o_count
    
    #avg purchase price by gender
        #Male
df_m = hero_df[hero_df["Gender"] == "Male"]
m_pp = df_m["Price"].mean()
        
        #Female
df_f = hero_df[hero_df["Gender"] == "Female"]
f_pp = df_f["Price"].mean()
        
        #Other
df_o = hero_df[hero_df["Gender"] == "Other / Non-Disclosed"]
o_pp = df_o["Price"].mean()
    
    #Average Purchase total by gender
        #male
m_ptot = df_m["Price"].sum()
        
        #female
f_ptot = df_f["Price"].sum()

        #other
o_ptot = df_o["Price"].sum()

    #average purchase total per person by gender
        #unique player dataframe for each gender
m_uniplyrs = df_m.groupby("SN")
f_uniplyrs = df_f.groupby("SN")
o_uniplyrs = df_o.groupby("SN")
        #male
m_unimean = m_uniplyrs["Price"].mean()

        #female
f_unimean = f_uniplyrs["Price"].mean()

        #other
o_unimean = o_uniplyrs["Price"].mean()
    #Create a summary data frame to hold the results
    
df_PAg = pd.DataFrame({"Purchase Count": [m_count, f_count, o_count], "Average Purchase Price": [m_pp, f_pp, o_pp], "Total Purchase Value": [m_ptot, f_ptot, o_ptot], "Average Total Purchase per Person": [m_unimean.mean(), f_unimean.mean(), o_unimean.mean()]})

    #Optional: give the displayed data cleaner formatting
    #Display the summary data frame
df_PAg
```
